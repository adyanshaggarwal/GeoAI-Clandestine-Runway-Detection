```python
#Code version 1
import os
import re
import cv2
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split
import torchvision.transforms.v2 as transforms
import matplotlib.pyplot as plt
from tqdm import tqdm
import warnings

warnings.filterwarnings("ignore")

# Install required packages for Google Colab
try:
    import torchvision.transforms.v2 as transforms
except ImportError:
    print("Installing torchvision...")
    !pip install torchvision>=0.15.0
    import torchvision.transforms.v2 as transforms

try:
    from PIL import Image
except ImportError:
    print("Installing Pillow...")
    !pip install Pillow
    from PIL import Image

# Optional: Install rasterio for better geospatial TIFF support
try:
    import rasterio
    print(f"Rasterio version: {rasterio.__version__}")
except ImportError:
    print("Installing rasterio for geospatial TIFF support...")
    !pip install rasterio
    import rasterio
    print(f"Rasterio version: {rasterio.__version__}")

from google.colab import drive
drive.mount('/content/drive')

# --- 1. Configuration & Setup ---
# TODO: Please update these paths to your data directories
IMAGE_DIR = '/content/drive/MyDrive/P2 - Amazon ITU - PESU/GEE_Exports_all'
MASK_DIR = '/content/drive/MyDrive/P2 - Amazon ITU - PESU/All_Masks_TIF'
MODEL_SAVE_PATH = '/content/drive/MyDrive/runway_segmentation_model_new.pth'

# Model Hyperparameters
LEARNING_RATE = 1e-4
BATCH_SIZE = 4
EPOCHS = 25
IMG_SIZE = 512

# Check for GPU
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {DEVICE}")


def test_image_readability(file_path):
    """Test if an image file can be read using multiple methods."""
    try:
        # Method 1: Rasterio (best for geospatial)
        import rasterio
        with rasterio.open(file_path) as src:
            # Just test opening and reading metadata
            print(f"    Rasterio success: {src.count} bands, {src.width}x{src.height}, dtype: {src.dtypes[0]}")
            return True

    except Exception as e:
        print(f"    Rasterio failed: {e}")

        # Method 2: OpenCV
        try:
            img = cv2.imread(file_path)
            if img is not None:
                print(f"    OpenCV success: shape {img.shape}")
                return True

            img = cv2.imread(file_path, cv2.IMREAD_UNCHANGED)
            if img is not None:
                print(f"    OpenCV (unchanged) success: shape {img.shape}")
                return True

        except Exception as e2:
            print(f"    OpenCV failed: {e2}")

        # Method 3: PIL
        try:
            from PIL import Image
            Image.MAX_IMAGE_PIXELS = None
            with Image.open(file_path) as pil_img:
                print(f"    PIL success: mode {pil_img.mode}, size {pil_img.size}")
                return True

        except Exception as e3:
            print(f"    PIL failed: {e3}")

    return False


def robust_image_load(file_path, target_mode='RGB'):
    """Load image using the most appropriate method for geospatial data."""
    try:
        # Method 1: Try rasterio first (best for geospatial TIFFs)
        import rasterio
        from rasterio.plot import reshape_as_image

        with rasterio.open(file_path) as src:
            # Read the image data
            img_data = src.read()  # Shape: (bands, height, width)

            if target_mode == 'RGB':
                # For RGB, we need 3 bands
                if img_data.shape[0] >= 3:
                    # Take first 3 bands and convert to (H, W, 3)
                    img_array = reshape_as_image(img_data[:3])
                    # Normalize to 0-255 if needed
                    if img_array.dtype != np.uint8:
                        # Normalize to 0-255 range
                        img_array = ((img_array - img_array.min()) /
                                   (img_array.max() - img_array.min()) * 255).astype(np.uint8)
                    return img_array
                else:
                    # If only 1 band, convert to RGB by repeating
                    single_band = reshape_as_image(img_data)
                    if single_band.dtype != np.uint8:
                        single_band = ((single_band - single_band.min()) /
                                     (single_band.max() - single_band.min()) * 255).astype(np.uint8)
                    return np.stack([single_band.squeeze()] * 3, axis=-1)

            elif target_mode == 'L':  # Grayscale
                # Take first band only
                img_array = img_data[0]  # Shape: (height, width)
                if img_array.dtype != np.uint8:
                    # Normalize to 0-255 range
                    img_array = ((img_array - img_array.min()) /
                               (img_array.max() - img_array.min()) * 255).astype(np.uint8)
                return img_array

    except Exception as e:
        print(f"Rasterio failed for {file_path}: {e}")

        # Method 2: Try GDAL through rasterio with different approach
        try:
            import rasterio
            with rasterio.open(file_path) as src:
                # Read as windowed read to handle large files
                window = rasterio.windows.Window(0, 0, src.width, src.height)
                img_data = src.read(window=window)

                if target_mode == 'RGB' and img_data.shape[0] >= 3:
                    # Take bands 1, 2, 3 (often RGB or NIR, Red, Green)
                    rgb_data = img_data[:3].transpose(1, 2, 0)  # Convert to H,W,C
                    # Normalize
                    for i in range(3):
                        band = rgb_data[:,:,i]
                        if band.max() > band.min():
                            rgb_data[:,:,i] = ((band - band.min()) / (band.max() - band.min()) * 255).astype(np.uint8)
                    return rgb_data.astype(np.uint8)

                elif target_mode == 'L':
                    # Use first band
                    gray_data = img_data[0]
                    if gray_data.max() > gray_data.min():
                        gray_data = ((gray_data - gray_data.min()) / (gray_data.max() - gray_data.min()) * 255).astype(np.uint8)
                    return gray_data.astype(np.uint8)

        except Exception as e2:
            print(f"Rasterio windowed read failed for {file_path}: {e2}")

        # Method 3: Try PIL with specific TIFF handling
        try:
            from PIL import Image
            Image.MAX_IMAGE_PIXELS = None  # Remove size limit

            with Image.open(file_path) as pil_img:
                # Handle multi-page TIFF
                if hasattr(pil_img, 'n_frames') and pil_img.n_frames > 1:
                    pil_img.seek(0)  # Go to first frame

                if target_mode == 'RGB':
                    # Try to convert to RGB
                    if pil_img.mode in ['RGB', 'RGBA']:
                        pil_img = pil_img.convert('RGB')
                    elif pil_img.mode in ['L', 'LA']:
                        # Convert grayscale to RGB
                        rgb_img = Image.new('RGB', pil_img.size)
                        if pil_img.mode == 'LA':
                            pil_img = pil_img.convert('L')
                        rgb_img.paste(pil_img)
                        rgb_img.paste(pil_img)
                        rgb_img.paste(pil_img)
                        pil_img = rgb_img
                    else:
                        pil_img = pil_img.convert('RGB')

                elif target_mode == 'L':
                    pil_img = pil_img.convert('L')

                img_array = np.array(pil_img)
                return img_array

        except Exception as e3:
            print(f"PIL with TIFF handling failed for {file_path}: {e3}")

        # Method 4: Fallback to OpenCV
        try:
            if target_mode == 'RGB':
                img = cv2.imread(file_path, cv2.IMREAD_COLOR)
                if img is not None:
                    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            else:  # Grayscale
                img = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
                if img is not None:
                    return img
        except Exception as e4:
            print(f"OpenCV also failed for {file_path}: {e4}")

    return None


def check_drive_access():
    """Check if Google Drive is properly mounted and accessible."""
    print("Checking Google Drive access...")

    # Check if drive is mounted
    if not os.path.exists('/content/drive'):
        print("Google Drive not mounted. Please run: drive.mount('/content/drive')")
        return False

    # Check if MyDrive exists
    if not os.path.exists('/content/drive/MyDrive'):
        print("MyDrive folder not accessible")
        return False

    print("Google Drive is accessible")

    # List contents of the project directory if it exists
    project_dir = '/content/drive/MyDrive/P2 - Amazon ITU - PESU'
    if os.path.exists(project_dir):
        print(f"Project directory found: {project_dir}")
        contents = os.listdir(project_dir)
        print(f"Contents: {contents}")
    else:
        print(f"Project directory not found: {project_dir}")
        print("Available directories in MyDrive:")
        try:
            mydrive_contents = os.listdir('/content/drive/MyDrive')
            for item in mydrive_contents[:10]:  # Show first 10 items
                print(f"  - {item}")
        except Exception as e:
            print(f"Error listing MyDrive contents: {e}")
        return False

    return True


# --- Enhanced Data Pairing Logic with Better Error Handling ---
def pair_image_mask_files(image_dir, mask_dir):
    """Pairs images with their corresponding masks based on filename."""
    # Check if directories exist
    if not os.path.exists(image_dir):
        raise FileNotFoundError(f"Image directory not found: {image_dir}")
    if not os.path.exists(mask_dir):
        raise FileNotFoundError(f"Mask directory not found: {mask_dir}")

    print(f"Checking image directory: {image_dir}")
    print(f"Checking mask directory: {mask_dir}")

    # Get all image and mask files
    all_image_files = os.listdir(image_dir)
    all_mask_files = os.listdir(mask_dir)

    print(f"Found {len(all_image_files)} files in image directory")
    print(f"Found {len(all_mask_files)} files in mask directory")

    # Filter for .tif files
    image_files = {f for f in all_image_files if f.lower().endswith(('.tif', '.tiff'))}
    mask_files = {f for f in all_mask_files if f.lower().endswith(('.tif', '.tiff'))}

    print(f"Found {len(image_files)} .tif image files")
    print(f"Found {len(mask_files)} .tif mask files")

    # Print some examples for debugging
    if image_files:
        print(f"Example image files: {list(image_files)[:5]}")
    if mask_files:
        print(f"Example mask files: {list(mask_files)[:5]}")

    paired_paths = []

    # Try multiple pairing strategies
    pairing_strategies = [
        # Strategy 1: Extract 'id_XXX.tif' from '0_id_XXX.tif' or '1_id_XXX.tif'
        (r'[01]_(id_\d+\.tif)', 1),
        # Strategy 2: More flexible pattern for id_numbers
        (r'[01]_(id_\d+)\.tif', lambda m: f"{m.group(1)}.tif"),
        # Strategy 3: Direct number matching
        (r'[01]_id_(\d+)\.tif', lambda m: f"id_{m.group(1)}.tif"),
    ]

    for strategy_num, (pattern, group_or_func) in enumerate(pairing_strategies, 1):
        print(f"\nTrying pairing strategy {strategy_num}: {pattern}")
        strategy_pairs = []

        for mask_name in mask_files:
            match = re.search(pattern, mask_name)
            if match:
                if callable(group_or_func):
                    image_name = group_or_func(match)
                else:
                    image_name = match.group(group_or_func)

                if image_name in image_files:
                    img_path = os.path.join(image_dir, image_name)
                    mask_path = os.path.join(mask_dir, mask_name)

                    # Verify files exist and are readable
                    if os.path.exists(img_path) and os.path.exists(mask_path):
                        # Test if files can be read with multiple methods
                        img_readable = test_image_readability(img_path)
                        mask_readable = test_image_readability(mask_path)

                        if img_readable and mask_readable:
                            strategy_pairs.append((img_path, mask_path))
                            print(f"  ✓ Paired: {mask_name} -> {image_name}")
                        else:
                            print(f"  ✗ Could not read: {mask_name} -> {image_name} (img: {img_readable}, mask: {mask_readable})")
                    else:
                        print(f"  ✗ Files don't exist: {mask_name} -> {image_name}")
                else:
                    print(f"  ✗ No matching image for mask: {mask_name} (looking for {image_name})")

        if strategy_pairs:
            paired_paths = strategy_pairs
            print(f"Strategy {strategy_num} successful! Found {len(paired_paths)} pairs.")
            break
        else:
            print(f"Strategy {strategy_num} found no pairs.")

    print(f"\nFinal result: Successfully paired {len(paired_paths)} image-mask files.")

    if len(paired_paths) == 0:
        print("\n" + "="*50)
        print("DEBUGGING INFORMATION")
        print("="*50)

        print(f"\nImage files sample (first 10):")
        for i, img_file in enumerate(sorted(list(image_files))[:10]):
            print(f"  {i+1}. {img_file}")

        print(f"\nMask files sample (first 10):")
        for i, mask_file in enumerate(sorted(list(mask_files))[:10]):
            print(f"  {i+1}. {mask_file}")

        print(f"\nTesting file readability...")
        # Test a few files to see if OpenCV can read them
        test_files = []
        if image_files:
            test_img_file = list(image_files)[0]
            test_img_path = os.path.join(image_dir, test_img_file)
            test_files.append(("Image", test_img_path))

        if mask_files:
            test_mask_file = list(mask_files)[0]
            test_mask_path = os.path.join(mask_dir, test_mask_file)
            test_files.append(("Mask", test_mask_path))

        for file_type, file_path in test_files:
            print(f"\nTesting {file_type}: {file_path}")
            print(f"  File exists: {os.path.exists(file_path)}")
            if os.path.exists(file_path):
                print(f"  File size: {os.path.getsize(file_path)} bytes")

                # Test multiple reading methods
                print("  Testing reading methods:")

                # OpenCV test
                test_read = cv2.imread(file_path)
                print(f"    OpenCV (default): {'Success' if test_read is not None else 'Failed'}")

                test_read2 = cv2.imread(file_path, cv2.IMREAD_UNCHANGED)
                print(f"    OpenCV (unchanged): {'Success' if test_read2 is not None else 'Failed'}")

                # PIL test
                try:
                    from PIL import Image
                    with Image.open(file_path) as pil_img:
                        print(f"    PIL: Success - Mode: {pil_img.mode}, Size: {pil_img.size}")
                except Exception as e:
                    print(f"    PIL: Failed - {e}")

                # File header check
                try:
                    with open(file_path, 'rb') as f:
                        header = f.read(10)
                        print(f"    File header: {header}")
                except Exception as e:
                    print(f"    Header read failed: {e}")

        print(f"\nPossible solutions:")
        print(f"1. Your TIFFs might be in a special format (GeoTIFF, multi-band, etc.)")
        print(f"2. Try installing additional packages: !pip install pillow rasterio")
        print(f"3. Files might be corrupted during upload")
        print(f"4. Try re-uploading with different compression settings")
        print("="*50)

    return paired_paths


# --- Enhanced Dataset Class with Better Error Handling ---
class RunwayDataset(Dataset):
    """Custom Dataset for loading runway images and masks."""
    def __init__(self, file_paths, img_size):
        self.file_paths = file_paths
        self.img_size = img_size

        # Updated transforms for compatibility
        self.transform = transforms.Compose([
            transforms.ToPILImage(),
            transforms.Resize((img_size, img_size)),
            transforms.ToImage(),
            transforms.ToDtype(torch.float32, scale=True),
        ])

        self.mask_transform = transforms.Compose([
            transforms.ToPILImage(),
            transforms.Resize((img_size, img_size), interpolation=transforms.InterpolationMode.NEAREST),
            transforms.ToImage(),
            transforms.ToDtype(torch.float32, scale=True),
        ])

    def __len__(self):
        return len(self.file_paths)

    def __getitem__(self, idx):
        img_path, mask_path = self.file_paths[idx]

        try:
            # Load image using robust method
            image = robust_image_load(img_path, target_mode='RGB')
            if image is None:
                raise ValueError(f"Could not load image: {img_path}")

            # Load mask using robust method
            mask = robust_image_load(mask_path, target_mode='L')
            if mask is None:
                raise ValueError(f"Could not load mask: {mask_path}")

            # Ensure mask is 2D
            if len(mask.shape) == 3:
                mask = mask[:, :, 0]  # Take first channel if multi-channel

            # Binarize mask: ensure it's 0 or 1
            mask = (mask > 127).astype(np.float32)

            # Add channel dimension to mask: (H, W) -> (H, W, 1)
            mask = np.expand_dims(mask, axis=-1)

            # Apply transforms
            image = self.transform(image)
            mask = self.mask_transform(mask)

            return image, mask

        except Exception as e:
            print(f"Error loading {img_path} or {mask_path}: {e}")
            # Return a dummy image and mask to prevent crashing
            dummy_image = torch.zeros((3, self.img_size, self.img_size))
            dummy_mask = torch.zeros((1, self.img_size, self.img_size))
            return dummy_image, dummy_mask


# --- U-Net Model (same as before) ---
class UNet(nn.Module):
    """A simplified U-Net model for segmentation."""
    def __init__(self, in_channels=3, out_channels=1):
        super(UNet, self).__init__()

        def double_conv(in_c, out_c):
            return nn.Sequential(
                nn.Conv2d(in_c, out_c, kernel_size=3, padding=1),
                nn.ReLU(inplace=True),
                nn.Conv2d(out_c, out_c, kernel_size=3, padding=1),
                nn.ReLU(inplace=True)
            )

        def up_conv(in_c, out_c):
            return nn.ConvTranspose2d(in_c, out_c, kernel_size=2, stride=2)

        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)

        # Downward path
        self.conv_down1 = double_conv(in_channels, 64)
        self.conv_down2 = double_conv(64, 128)
        self.conv_down3 = double_conv(128, 256)

        # Upward path
        self.up_trans2 = up_conv(256, 128)
        self.conv_up2 = double_conv(256, 128)

        self.up_trans1 = up_conv(128, 64)
        self.conv_up1 = double_conv(128, 64)

        # Output layer
        self.out = nn.Conv2d(64, out_channels, kernel_size=1)

    def forward(self, x):
        # Down
        x1 = self.conv_down1(x)
        x2 = self.pool(x1)
        x3 = self.conv_down2(x2)
        x4 = self.pool(x3)
        x5 = self.conv_down3(x4)

        # Up
        x = self.up_trans2(x5)
        x = self.conv_up2(torch.cat([x3, x], dim=1))

        x = self.up_trans1(x)
        x = self.conv_up1(torch.cat([x1, x], dim=1))

        return self.out(x)


# --- Training and Validation Functions (same as before) ---
def train_fn(loader, model, optimizer, loss_fn, device):
    """A single training epoch."""
    model.train()
    loop = tqdm(loader, desc="Training")
    running_loss = 0.0
    for batch_idx, (data, targets) in enumerate(loop):
        data = data.to(device)
        targets = targets.to(device)

        # Forward pass
        predictions = model(data)
        loss = loss_fn(predictions, targets)

        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        loop.set_postfix(loss=loss.item())

    return running_loss / len(loader)

def validate_fn(loader, model, loss_fn, device):
    """Validation loop."""
    model.eval()
    val_loss = 0.0
    with torch.no_grad():
        for data, targets in loader:
            data = data.to(device)
            targets = targets.to(device)
            predictions = model(data)
            val_loss += loss_fn(predictions, targets).item()

    return val_loss / len(loader)

# --- Enhanced Main Function ---
def main():
    print("Starting runway segmentation training...")

    # Check Google Drive access first
    if not check_drive_access():
        print("Please fix Google Drive access issues before continuing.")
        return

    # Setup and validation
    try:
        paired_files = pair_image_mask_files(IMAGE_DIR, MASK_DIR)
    except FileNotFoundError as e:
        print(f"Error: {e}")
        print("Please check your directory paths in the configuration section.")
        return

    if not paired_files:
        print("No paired files found. Please check your directories and naming conventions.")
        print(f"Looking for images in: {IMAGE_DIR}")
        print(f"Looking for masks in: {MASK_DIR}")
        return

    print(f"Creating dataset with {len(paired_files)} image-mask pairs...")
    dataset = RunwayDataset(paired_files, img_size=IMG_SIZE)

    # Split data: 80% for training, 20% for validation
    train_size = int(0.8 * len(dataset))
    val_size = len(dataset) - train_size
    train_dataset, val_dataset = random_split(dataset, [train_size, val_size])

    print(f"Training samples: {len(train_dataset)}, Validation samples: {len(val_dataset)}")

    # Create DataLoaders
    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2)
    val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2)

    # Model, Loss, Optimizer
    print("Initializing model...")
    model = UNet(in_channels=3, out_channels=1).to(DEVICE)
    loss_fn = nn.BCEWithLogitsLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=LEARNING_RATE)

    print(f"Model has {sum(p.numel() for p in model.parameters()):,} parameters")

    # Training loop
    history = {'train_loss': [], 'val_loss': []}
    best_val_loss = float('inf')

    print(f"\nStarting training for {EPOCHS} epochs...")
    for epoch in range(EPOCHS):
        print(f"\n--- Epoch {epoch+1}/{EPOCHS} ---")

        try:
            train_loss = train_fn(train_loader, model, optimizer, loss_fn, DEVICE)
            val_loss = validate_fn(val_loader, model, loss_fn, DEVICE)

            print(f"Epoch {epoch+1} -> Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}")

            history['train_loss'].append(train_loss)
            history['val_loss'].append(val_loss)

            # Save the best model
            if val_loss < best_val_loss:
                best_val_loss = val_loss
                torch.save(model.state_dict(), MODEL_SAVE_PATH)
                print(f"New best model saved! (Val Loss: {val_loss:.4f})")

        except Exception as e:
            print(f"Error during epoch {epoch+1}: {e}")
            break

    print("\nTraining completed!")
    print(f"Best validation loss: {best_val_loss:.4f}")

    # Visualization
    try:
        plot_training_results(history)
        visualize_predictions(val_loader, model, DEVICE)
    except Exception as e:
        print(f"Error during visualization: {e}")

# --- Plotting and Visualization Functions ---
def plot_training_results(history):
    """Plots training and validation loss."""
    plt.figure(figsize=(10, 5))
    plt.plot(history['train_loss'], label='Training Loss', marker='o')
    plt.plot(history['val_loss'], label='Validation Loss', marker='s')
    plt.title('Training and Validation Loss Over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Loss (BCEWithLogitsLoss)')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

def visualize_predictions(loader, model, device, num_samples=3):
    """Displays original image, true mask, and predicted mask."""
    model.eval()

    try:
        images, masks = next(iter(loader))
        images, masks = images.to(device), masks.to(device)

        with torch.no_grad():
            preds = torch.sigmoid(model(images))  # Apply sigmoid to get 0-1 range
            preds = (preds > 0.5).float()         # Binarize predictions

        # Move to CPU for plotting
        images = images.cpu().permute(0, 2, 3, 1).numpy()
        masks = masks.cpu().permute(0, 2, 3, 1).numpy()
        preds = preds.cpu().permute(0, 2, 3, 1).numpy()

        plt.figure(figsize=(15, 5 * num_samples))
        plt.suptitle("Model Predictions vs. Ground Truth", fontsize=16)

        for i in range(min(num_samples, len(images))):
            # Original Image
            plt.subplot(num_samples, 3, i * 3 + 1)
            plt.imshow(images[i])
            plt.title("Original Image")
            plt.axis('off')

            # True Mask
            plt.subplot(num_samples, 3, i * 3 + 2)
            plt.imshow(masks[i].squeeze(), cmap='gray')
            plt.title("True Mask")
            plt.axis('off')

            # Predicted Mask
            plt.subplot(num_samples, 3, i * 3 + 3)
            plt.imshow(preds[i].squeeze(), cmap='gray')
            plt.title("Predicted Mask")
            plt.axis('off')

        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        plt.show()

    except Exception as e:
        print(f"Error in visualization: {e}")


# --- Installation Check Function ---
def check_requirements():
    """Check if all required packages are installed."""
    print("Checking requirements...")

    # Check PyTorch
    print(f"PyTorch version: {torch.__version__}")
    print(f"CUDA available: {torch.cuda.is_available()}")
    if torch.cuda.is_available():
        print(f"CUDA device: {torch.cuda.get_device_name()}")

    # Check OpenCV
    print(f"OpenCV version: {cv2.__version__}")

    try:
        import torchvision.transforms.v2 as transforms_v2
        print("torchvision v2 transforms: Available")
    except ImportError:
        print("torchvision v2 transforms: Not available - installing...")
        get_ipython().system('pip install torchvision>=0.15.0')


if __name__ == '__main__':
    # Check requirements first
    check_requirements()

    # Run main function with error handling
    try:
        main()
    except KeyboardInterrupt:
        print("\nTraining interrupted by user.")
    except Exception as e:
        print(f"\nAn unexpected error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import re
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split
import torchvision.transforms.v2 as transforms
import matplotlib.pyplot as plt
from tqdm import tqdm
import warnings
import rasterio
from rasterio.plot import reshape_as_image

warnings.filterwarnings("ignore")

# --- 1. Configuration & Setup ---
# TODO: UPDATE THESE PATHS to your data directories on Google Drive!
IMAGE_DIR = '/content/drive/MyDrive/P2 - Amazon ITU - PESU/GEE_Exports_all'
MASK_DIR = '/content/drive/MyDrive/P2 - Amazon ITU - PESU/All_Masks_TIF'
MODEL_SAVE_PATH = '/content/drive/MyDrive/runway_segmentation_model_new.pth'

# Stable Hyperparameters
LEARNING_RATE = 1e-4
BATCH_SIZE = 4  # Keep this low to prevent memory issues
EPOCHS = 25
IMG_SIZE = 256  # Start with 256 for stability

# Check for GPU
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {DEVICE}")


# --- 2. Data Pairing Logic (Unchanged) ---
def pair_image_mask_files(image_dir, mask_dir):
    """Pairs images with their corresponding masks."""
    try:
        image_files = {f for f in os.listdir(image_dir) if f.endswith('.tif')}
        mask_files = {f for f in os.listdir(mask_dir) if f.endswith('.tif')}
    except FileNotFoundError as e:
        print(f"ERROR: A data directory was not found. Please double-check your paths.")
        print(e)
        return []

    paired_paths = []
    for mask_name in mask_files:
        match = re.search(r'\d_(id_\w+\.tif)', mask_name)
        if match:
            image_name = match.group(1)
            if image_name in image_files:
                img_path = os.path.join(image_dir, image_name)
                mask_path = os.path.join(mask_dir, mask_name)
                paired_paths.append((img_path, mask_path))

    print(f"Found {len(paired_paths)} paired image-mask files.")
    return paired_paths

# --- 3. Robust `rasterio` Image Loader ---
def load_with_rasterio(path, target_mode='RGB'):
    """Loads a geospatial TIFF file and converts it to a standard NumPy array."""
    with rasterio.open(path) as src:
        # Read the bands. This might be float data, not uint8.
        img_data = src.read()

    # Handle images with more than 3 bands (e.g., RGB + NIR) by taking the first 3
    if target_mode == 'RGB':
        if img_data.shape[0] >= 3:
            img_data = img_data[:3]
        # If it's a single-band image, stack it to create an RGB-like image
        elif img_data.shape[0] == 1:
            img_data = np.vstack([img_data[0], img_data[0], img_data[0]])

    # For masks, just take the first band
    elif target_mode == 'L':
        img_data = img_data[[0], :, :]

    # Convert from (C, H, W) to (H, W, C) for image processing
    img_array = reshape_as_image(img_data)

    # Normalize pixel values to the 0-255 range expected by image libraries
    # This is crucial for satellite data which is often not in uint8 format
    img_array = np.nan_to_num(img_array) # Remove any NaN values
    min_val, max_val = np.min(img_array), np.max(img_array)

    if max_val > min_val:
        # Stretch the pixel values to fill the 0-255 range
        img_array = ((img_array - min_val) / (max_val - min_val) * 255).astype(np.uint8)
    else:
        # Handle cases where the image is all one color
        img_array = np.zeros_like(img_array, dtype=np.uint8)

    return img_array


# --- 4. Custom Dataset using `rasterio` ---
class RunwayDataset(Dataset):
    """Dataset that uses the robust `rasterio` loader."""
    def __init__(self, file_paths, img_size):
        self.file_paths = file_paths
        self.img_size = img_size
        self.transform = transforms.Compose([
            transforms.ToPILImage(),
            transforms.Resize((img_size, img_size)),
            transforms.ToTensor(),
        ])
        self.mask_transform = transforms.Compose([
            transforms.ToPILImage(),
            transforms.Resize((img_size, img_size), interpolation=transforms.InterpolationMode.NEAREST),
            transforms.ToTensor(),
        ])

    def __len__(self):
        return len(self.file_paths)

    def __getitem__(self, idx):
        img_path, mask_path = self.file_paths[idx]

        try:
            image = load_with_rasterio(img_path, target_mode='RGB')
            mask = load_with_rasterio(mask_path, target_mode='L')
        except Exception as e:
            # If loading fails, raise an error with the file path
            raise IOError(f"Failed to load file: {img_path} or {mask_path}. Reason: {e}")

        # Binarize mask to ensure it's just 0s and 1s
        mask = (mask > 127).astype(np.uint8) * 255

        # Apply transforms
        image = self.transform(image)
        mask = self.mask_transform(mask)

        return image, mask

# --- Simple U-Net Model and all other functions are the same as the previous stable script ---
# (They are included here for completeness)

class UNet(nn.Module):
    """A simplified U-Net model, sufficient for most segmentation tasks."""
    def __init__(self, in_channels=3, out_channels=1):
        super(UNet, self).__init__()

        def double_conv(in_c, out_c):
            return nn.Sequential(
                nn.Conv2d(in_c, out_c, kernel_size=3, padding=1),
                nn.ReLU(inplace=True),
                nn.Conv2d(out_c, out_c, kernel_size=3, padding=1),
                nn.ReLU(inplace=True)
            )

        def up_conv(in_c, out_c):
            return nn.ConvTranspose2d(in_c, out_c, kernel_size=2, stride=2)

        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        self.conv_down1 = double_conv(in_channels, 64)
        self.conv_down2 = double_conv(64, 128)
        self.up_trans1 = up_conv(128, 64)
        self.conv_up1 = double_conv(128, 64)
        self.out = nn.Conv2d(64, out_channels, kernel_size=1)

    def forward(self, x):
        # Down
        x1 = self.conv_down1(x)
        x2 = self.pool(x1)
        x3 = self.conv_down2(x2)
        # Up
        x = self.up_trans1(x3)
        x = self.conv_up1(torch.cat([x1, x], dim=1))
        return self.out(x)

def main():
    paired_files = pair_image_mask_files(IMAGE_DIR, MASK_DIR)
    if not paired_files:
        print("Stopping execution because no data files were paired.")
        return

    dataset = RunwayDataset(paired_files, img_size=IMG_SIZE)
    train_size = int(0.8 * len(dataset))
    val_size = len(dataset) - train_size
    train_dataset, val_dataset = random_split(dataset, [train_size, val_size])

    print(f"Training samples: {len(train_dataset)}, Validation samples: {len(val_dataset)}")

    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)

    model = UNet(in_channels=3, out_channels=1).to(DEVICE)
    loss_fn = nn.BCEWithLogitsLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=LEARNING_RATE)

    history = {'train_loss': [], 'val_loss': []}
    best_val_loss = float('inf')

    for epoch in range(EPOCHS):
        print(f"\n--- Epoch {epoch+1}/{EPOCHS} ---")
        model.train()
        train_loss = 0.0
        loop = tqdm(train_loader, desc="Training")
        for data, targets in loop:
            data, targets = data.to(DEVICE), targets.to(DEVICE)
            optimizer.zero_grad()
            predictions = model(data)
            loss = loss_fn(predictions, targets)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()
            loop.set_postfix(loss=loss.item())
        avg_train_loss = train_loss / len(train_loader)
        history['train_loss'].append(avg_train_loss)

        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for data, targets in val_loader:
                data, targets = data.to(DEVICE), targets.to(DEVICE)
                predictions = model(data)
                val_loss += loss_fn(predictions, targets).item()
        avg_val_loss = val_loss / len(val_loader)
        history['val_loss'].append(avg_val_loss)

        print(f"Epoch {epoch+1} -> Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}")

        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            torch.save(model.state_dict(), MODEL_SAVE_PATH)
            print(f" Model saved to {MODEL_SAVE_PATH}")

    plot_training_results(history)
    print("\nLoading best model for visualization...")
    model.load_state_dict(torch.load(MODEL_SAVE_PATH))
    visualize_predictions(val_loader, model, DEVICE)

def plot_training_results(history):
    plt.figure(figsize=(10, 5))
    plt.plot(history['train_loss'], label='Training Loss')
    plt.plot(history['val_loss'], label='Validation Loss')
    plt.title('Training and Validation Loss Over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()

def visualize_predictions(loader, model, device, num_samples=5):
    model.eval()
    images, masks = next(iter(loader))
    images, masks = images.to(device), masks.to(device)

    with torch.no_grad():
        preds = torch.sigmoid(model(images))
        preds_binary = (preds > 0.5).float()

    images = images.cpu().permute(0, 2, 3, 1)
    masks = masks.cpu().permute(0, 2, 3, 1).squeeze()
    preds_binary = preds_binary.cpu().permute(0, 2, 3, 1).squeeze()

    plt.figure(figsize=(15, 5 * num_samples))
    plt.suptitle("Model Predictions vs. Ground Truth", fontsize=16)
    for i in range(min(num_samples, len(images))):
        plt.subplot(num_samples, 3, i * 3 + 1)
        plt.imshow(images[i])
        plt.title("Original Image")
        plt.axis('off')
        plt.subplot(num_samples, 3, i * 3 + 2)
        plt.imshow(masks[i], cmap='gray')
        plt.title("True Mask")
        plt.axis('off')
        plt.subplot(num_samples, 3, i * 3 + 3)
        plt.imshow(preds_binary[i], cmap='gray')
        plt.title("Predicted Mask")
        plt.axis('off')
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()


if __name__ == '__main__':
    try:
        main()
    except Exception as e:
        print(f"\n An unexpected error occurred: {e}")
        print("\nDebugging suggestions:")
        print("1. Double-check your IMAGE_DIR and MASK_DIR paths.")
        print("2. Ensure your Colab runtime has a GPU enabled.")
        print("3. If you get a 'CUDA out of memory' error, reduce BATCH_SIZE.")
```
