Imports and Mounting

```python
from google.colab import drive
import os
import shutil
```

```python
# Mount Google Drive
drive.mount('/content/drive', force_remount=True)
```

Change source and destination file paths below according to your use-case.

```python
# Set folder paths - change these to your actual Drive folder paths
source_folder = '/content/drive/MyDrive/Annotated images/tiff'
destination_folder = '/content/drive/MyDrive/Sentinel_training'

# Starting ID based on last existing file in destination (last was 182)
starting_id = 235

# List all .tif files in source folder
source_files = sorted([f for f in os.listdir(source_folder) if f.endswith('.tif')])

# Move and rename files
for i, filename in enumerate(source_files):
    new_id = starting_id + i
    new_filename = f"Sentinel_AllBands_Training_Id_{new_id}.tif"

    source_path = os.path.join(source_folder, filename)
    dest_path = os.path.join(destination_folder, new_filename)

    shutil.move(source_path, dest_path)

print(f"Moved and renamed {len(source_files)} files.")
```
